# Capacitor

## Introduction

**Electrical capacitance** represents the ability of a component to **store energy in the form of an electric charge**.
It is measured in **farads (F)** and defined as the **ratio between the stored charge** on the plates of a capacitor and the **voltage applied** across them.
The main device that exhibits this property is the **capacitor** itself.
